```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def distribuir_pontos(x1, y1, x2, y2, N):
    x = np.linspace(x1, x2, N)
    y = np.linspace(y1, y2, N)
    return x, y
```
